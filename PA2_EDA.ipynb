##Data Analysis & Feature Engineering

#### List of categorical and numerical features

```python
# Excluding the labels column, i.e., PCOS:
not_to_be_used = ['Sl. No', 
                  'Patient File No.', 
                  'PCOS (Y/N)']
labels = 'PCOS (Y/N)'

# Get features
features_categ = []
features_numeric = []
for column in PCOS.columns:
  unq = PCOS[column].unique()
  if len(unq) <= 3 and (column not in not_to_be_used):
    # print(f"{len(unq)} \t {column}: {unq}" )
    features_categ.append(column)

features_categ += ['Blood Group']


for column in PCOS.columns:
  if column not in features_categ and (column not in not_to_be_used):
    features_numeric.append(column)


features_numeric += ['Cycle(R/I)']

print(f"Labels: {labels}\n")
print(f"Categorical: {features_categ}\n")
print(f"Numerical: {features_numeric}\n")

print(f"Total of features: {len(features_categ) + len(features_numeric)}")
```

####Issues in the data:

```python
# Missing values
# NaNs
missing_vals_cols = ['Marraige Status (Yrs)', 'Fast food (Y/N)']

rows_miss_vals = []
for col in missing_vals_cols:
  row = PCOS[col].index[PCOS[col].isnull()]
  rows_miss_vals.append(row[0])


print(f"{'---'*20}")
print(f"Features with nan values:\n{missing_vals_cols}\n")
print(f"Indices of examples with nan values:\n{rows_miss_vals}")
print(f"{'---'*20}")
```

```python
# Missing vals
# Check for off values, such as '?', or any other kind of characters
print(PCOS.dtypes)

# There are "odd" values in the features "II    beta-HCG(mIU/mL)", and \
# "AMH(ng/mL)"
# The following will return an error:
# data_part1["II    beta-HCG(mIU/mL)"].copy().astype("float64")
# data_part1["AMH(ng/mL)"].copy().astype("float64")

def get_str(val):
  if type(val)==str:
    return 1

a = PCOS["AMH(ng/mL)"].apply(get_str)
a_idx = a.index[a==1]

b = PCOS["II    beta-HCG(mIU/mL)"].apply(get_str)
b_idx = b.index[b==1]


print(f"\nFeature 'AMH(ng/mL)' has a none float64 value in row: {a_idx[0]}\n")
print(f"Feature 'II    beta-HCG(mIU/mL)' has a none float64 value in row: {b_idx[0]}\n")
```

We will transform 'II    beta-HCG(mIU/mL)' and 'AMH(ng/mL)' from a oject type to a float type. This will be done in the univariate analysis
