## **Some that required further description based on our meeting:**
- **'  I   beta-HCG(mIU/mL)' and 'II    beta-HCG(mIU/mL)'**: human chorionic gonadotropin (HCG). A quantitative test that determines the levels of the HCG hormone in the blood. The HCG hormone is produced in the body during pregancy (HCG Blood Test - Quantitative, n.d.).
- **FSH**: Follicle-stimulating hormone.
- **LH**: Luteinizing hormone. Useful in the monitoring of ovulation, infertility, pituitary disorders, among others (Luteinizing Hormone (LH), Serum. (n.d.)).
- **AMH**: anti-mullerian hormone. “healthcare providers frequently use the test in women — mainly to provide a snapshot of a woman’s reproductive health.” (Anti-Mullerian Hormone Test. (n.d.)).
- **TSH**: thyroid stimulating homone.
- **PRL**: Prolactin blood test. Prolactin stimulates breast development and milk production in women (Prolactin blood test. (n.d.)).
- **PRG**: 17-OH progesterone. “The main use of this test is to check infants for an inherited disorder that affects the adrenal gland, called congenital adrenal hyperplasia (CAH). It is often done on infants who are born with outer genitals that do not clearly look like those of a boy or a girl.” (17-OH progesterone. (n.d.)., Congenital adrenal hyperplasia. (n.d.)).
RBS: Blood sugar levels.

# **Performance, time, and cost.**

## 1. **Performance**

There are several tests to quantify the performance of a machine learning model. Some of those are accuracy, precision, sensitivity and specificity. In the health care sector sensitivity and specificity are broadly used. For this project, sensitivity will tell how good the model is for classifying women with PCOS. On the other hand, specificity will inform how good the model is for classifying women that do not have PCOS. These metrics provide a score from 0 to 1 or from 0 to 100, being 1 (100) the best level of classification. It is important to note that these scores give information about the performance of the model for a specific threshold that will classify women with and without PCOS. If the data of this study is imbalanced, the area under the curve (AUC) score could help for this potential issue. AUC score will summarize the metrics at all PCOS classification thresholds. The ideal scenario will be a performance with metrics with more than 90%. For instance, a model that applied the random forest algorithm for PCOS prediction showed 93.25 % of accuracy (Tiwari, et al., 2022).

**Corrective actions to our ML model:**

- A key approach to properly evaluating the model’s performance could be to examine how well the trained model responds to new unseen data often known as cross validation. We could achieve this by splitting the data set using a 70-30 percent approach with 70% of the data allocated for training and the remaining 30% for testing. A type of cost function approach known as the cross entropy function will be used to evaluate the difference between the predicted result and the actual value. The lesser the cost or difference (loss) is, the more accurate the model is which translates to the best model. The model will be run iteratively to find parameters, relationships and features that reduce the cost function.

- Different evaluation metrics can be explored, such as the confusion matrix which we can use to calculate the precision and recall rate since our PCOS prediction model is a binary classification type. The matrix classifies the result as a True Positive, True Negative, False Positive, and False Negative.

- In order to correct the amount of false results, a proper analysis of both the data set and model algorithm could be examined to determine the error source and corrective actions such as optimizing bias and variance (overfitting and underfitting), re-evaluating the selected features , and fine tuning the parameters in the algorithm will be implemented.

# **Ethical description**

There are five circumstances to consider in clinical research for ethical purposes; these are privacy, fairness, accountability, and transparency (Basu, et al., 2020).

- **Privacy**
Data privacy from medical electronic records has specific regulations depending on which region or country the study is developed (Basu, et al., 2020). Data of this machine learning project comes from different hospitals in the state of Kerala, in India (Kaggle, n.d). Thus, in this case, data was under the regulation of the Indian Medical Council Act (Basu, et al., 2020), which defines “Defining privacy between patient and doctor” (Basu, et al., 2020). Another ethical guideline for this data was the Digital Information Security in Healthcare Act (Basu, et al., 2020), which “ensures electronic health data privacy, confidentiality, security and standardization” (Basu, et al., 2020). In order to keep the confidentiality of the participants of this data, the kaggle author did not include personal information such as name, address, health care identification or in which hospital the clinical information was taken. Therefore, it is safe to say that there is not an ethical issue for privacy in this case.

- **Fairness**
This refers to the potential bias in the acquisition of the data (Basu, et al., 2020). For this machine learning project it is not clear if the quantification of the different medical tests were carried out with the same technology. Also, it is unknown if there is a diverse set of participants that include women from different areas in Kerala or from different socioeconomic backgrounds. In summary, there is uncertainty for the test measurements and the demographic information of the participants.

- **Accountability**
In ethics, accountability is the circumstance where a person, a group of people or an organization takes responsibility for particular outcomes by accepting moral and legal actions (Basu, et al., 2020). Artificial intelligence and machine learning creates a gap for accountability. For instance, at some point, the developer of a machine learning technology could not be able to predict “the future machine behavior anymore and thus cannot be held morally responsible or liable for it” (Basu, et al., 2020). It seems that for a classification PCOS model there is not a future unpredictable behavior beyond of determining which patient has or do not have a PCOS

- **Transparency**
Users and developers of the PCOS machine learning classification model have the ethical responsibility of  reporting the most suitable metrics (Basu, et al., 2020). For example, in the case of imbalance data, accuracy will not be the best way to report the performance of the model because this metric could mislead the diagnosis of PCOS.
